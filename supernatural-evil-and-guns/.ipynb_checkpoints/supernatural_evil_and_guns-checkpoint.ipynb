## The Data

Our analysis is based on the data collected in 2014 from the Wave Four of the Baylor Religion Survey (n=1572), which includes answers to various questions regarding the values and beliefs of people living in the US. It covers topics of religious behaviors and attitudes; morality, gender roles, and politics; family and religiosity; sexual orientation; work; race and ethnicity; guns and society; surveillance; science and the supernatural; and basic demographics.

We mainly focused on the religious bahaviours and attitudes, politics, race and ethnicity, guns and society, and the belief in the supernatural.

Firstly , we wanted to investigate whether there is a connection between the belief in the supernatural evil and the opinion people have on various policies such as a ban on semi-auto guns, ban on high-capacity ammo clips, ban on civilian bandguns and others.

In order to measure the belief in supernatural evil, we will use the answers to three questions asked by the participants in the survey:

* Whether the respondent believes in the devil.

* Whether the respondent believes in hell.

* Whether the respondent believes in demons.

The metric we're going to use in order to measure the belief in superantural evil is calculated by summing the responses given in each of the above questions and then averaged across the number of items answered by the participant.

```python
supernatural = data.loc[:,["Q23A", "Q23C", "Q23G"]]
supernatural.rename(columns={'Q23A': 'devil', 'Q23C': 'hell', 'Q23G': 'demons'}, inplace=True)
supernatural["sum"] = supernatural.sum(axis=1,min_count=1)
supernatural["supernatural_evil_metric"] = supernatural["sum"] / (3 - supernatural.loc[:,["devil", "hell", "demons"]].isna().sum(axis=1))
supernatural
```

This metric ranges from 1 to 4 with a mean of 3.09, which shows significant levels of overall belief in supernatural evil among the respondents.

```python
supernatural.supernatural_evil_metric.mean()
```

## Decision Trees and Random Forest

Now, we'll continue by carrying out predictions using Decision Trees and Random Forests and then choose the models works best for our dataset. Let's see whether the belief in supernatural evil comes up as a stronger predictor here.

```python
features = df.iloc[:, :24]
policies = df.iloc[:, 24:]
```

* First, we'll find the best depth for each decision tree regarding a gun policy.

```python
best_max_depth = []
for i in range(8):
    parameters = [
      {'max_depth': list(range(1, 11))},
     ]

    cv = KFold(n_splits=5, shuffle=True, random_state=13)
    clf = GridSearchCV(DecisionTreeRegressor(), parameters, cv=cv)

    clf = clf.fit(features, policies.iloc[:,i])
    print(clf.best_estimator_)
    best_max_depth.append(clf.best_params_)
    print(policies.columns[i], clf.best_params_)
```

* Now, we're going to run a Decision Tree Classifier and a Bagging Classifier and compare their results on their predictions.

### Decision Tree Classifier

```python
for i in range(8):
    tree = DecisionTreeClassifier(random_state=0, criterion='entropy')
    cv_score = cross_val_score(tree, 
                               features, 
                               policies.iloc[:,i], cv=10,
                               scoring='accuracy')

    print(policies.columns[i],": ", np.mean(cv_score), sem(cv_score))
```

### Bagging Classifier

```python
for i in range(8):
    bagging_tree = BaggingClassifier(DecisionTreeClassifier(criterion='entropy'),
                                           n_estimators=50, 
                                           n_jobs=-1)
    bagged_cv_score = cross_val_score(bagging_tree, 
                                      features, 
                                      policies.iloc[:,i], 
                                      cv=3, scoring='accuracy')

    print(policies.columns[i],": ", np.mean(bagged_cv_score), ", " , sem(bagged_cv_score))
```

By looking at the mean validation score along with the standard error of the mean, we can see than the Bagging Classifier work way better that the Decision Tree Classifier, but now let's try a Boosting method. 

* We'll run an AdaBoost Classifier and compare the results once again.

### AdaBoost Classifier

```python
for i in range(8):
    clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1),n_estimators=150)
    scores = cross_val_score(clf, features, 
                             policies.iloc[:,i], cv=5)
    print(policies.columns[i], ": ", scores.mean(), ", " , sem(scores))
```

Here, the AdaBoost Classifier worked even better.

Now let's see if we can get better predictions using Random Forests or Extremely Randomized Trees. 

* We'll run a Random Forest Classifier and a Extra Trees Classifier and compare the results.

### Random Forest Classifier

```python
for i in range(8):
    forest = RandomForestClassifier(n_estimators=50, max_depth=None,
                                    min_samples_split=2)
    scores = cross_val_score(forest, features, 
                             policies.iloc[:,i], cv=5)
    print(policies.columns[i],": ", scores.mean(), ", " , sem(scores))
```

### Extra Trees Classifier

```python
for i in range(8):
    extremely_rts = ExtraTreesClassifier(n_estimators=50, 
                                         max_depth=None,
                                         min_samples_split=2)

    scores = cross_val_score(extremely_rts, features, policies.iloc[:,i], cv=5)
    print(policies.columns[i],": ", scores.mean(), ", " , sem(scores))
```

Comparing all the methods, we can see that the Random Forest Classifier worked the best in predicting the support of the different gun policies. So, let's see the feature importances for each gun policy prediction.
